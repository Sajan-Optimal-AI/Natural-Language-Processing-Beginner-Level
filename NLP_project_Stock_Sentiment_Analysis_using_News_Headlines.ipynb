```python
# Import the data
import pandas as pd
Data = pd.read_csv('/content/drive/MyDrive/News headline for stock sentiment NLP.csv', encoding = "ISO-8859-1")  # if dataset contains non-ascii characters use encoding
Data
```

```python
Data.shape
```

```python
# Split train and test data 
# Date feature ranges between years 2000 to 2016..So instances till the end of 2014 is added in train data and instances at the year 2015 and 2016 is added in test data.
Train_data_Raw = Data[Data['Date']<'20150101']        # 2015-01-01
Test_data_Raw = Data[Data['Date']>'20141231']         # 2014-12-31
print(Test_data_Raw.shape,Train_data_Raw.shape)
```

```python
# Text Cleaning and preprocessing train_data:
Train_Data = Train_data_Raw.iloc[:,2:27]
Train_Data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)        # regex is regular expression(re)
Test_Data = Test_data_Raw.iloc[:,2:27]
Test_Data.replace("[^a-zA-Z]"," ",regex=True, inplace=True) 
```

```python
 #Renaming training data columns 
list1 = [i for i in range(25)]
new_index_Train = [str(i) for i in list1]
Train_Data.columns = new_index_Train
Train_Data.head()
```

```python
# Renaming testing data columns
list2 = [i for i in range(25)]
new_index_Test = [str(i) for i in list2]
Test_Data.columns = new_index_Test
Test_Data.head()
```

```python
# converting all the headlines into lowercase
for i in new_index:
  Train_Data[i] = Train_Data[i].str.lower()
Train_Data.head()
```

```python
# converting all the headlines into lowercase
for i in new_index:
  Test_Data[i] = Test_Data[i].str.lower()
Test_Data.head()
```

```python
# Appending all the headlines:(Train_Data)
Train_headlines = []
for row in range(0,len(Train_Data.index)):
  Train_headlines.append(' '.join(str(x) for x in Train_Data.iloc[row,0:25]))
Train_headlines[5]    #  fifth index or row 
```

```python
#Appending all the headlines: (Test_Data)
Test_headlines = []
for row in range(0,len(Test_Data.index)):
  Test_headlines.append(' '.join(str(x) for x in Test_Data.iloc[row,0:25]))
```

```python
# Define the model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
# implement TF-IDF on Train_headlines
TF_IDF_T =TfidfVectorizer(max_features = 15484)
Train_dataset=TF_IDF_T.fit_transform(Train_headlines)
Train_dataset.shape
```

```python
# implement TF-IDF on Test_headlines
TF_IDF_Te =TfidfVectorizer()
Test_dataset=TF_IDF_Te.fit_transform(Test_headlines)
Test_dataset.shape
```

```python
#Train the model
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(Train_dataset,Train_data_Raw['Label'])
```

```python
# Evaluate the model
predictions = randomclassifier.predict(Test_dataset) 
(predictions)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
matrix=confusion_matrix(Test_data_Raw['Label'],predictions)
print(matrix)
score=accuracy_score(Test_data_Raw['Label'],predictions)
print(score)
report=classification_report(Test_data_Raw['Label'],predictions)
print(report)
```
